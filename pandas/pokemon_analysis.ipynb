```python
import pandas as pd

# loading a csv as a dataframe, which is a data-structure provided by pandas to query complex data
df = pd.read_csv('pokemon_data.csv')
```

# ways to Iterate over the dataframe 

```python

df.head() #value of top of the df(default 5 rows, can be sepcified as args)
df.tail() #value of bottom of the df(default 5 rows, can be sepcified as args)
df.columns #value of columns names
df[['Name', 'Type 1']][0:5] #gives the data of specified column name
df.Name[0:5] # same as above, but different syntax

df.iloc[1:4] # read each rows as specified

df.iloc[0,2] # value of what is on row:1 and column:2


# df.iterrows() returns with (index,series) pairs where series is content of row at index(series will be a dictionary)
# for index, row in df.iterrows():
#     print(index, row)
#     type(row)


# df.loc() this also helps us to locate data of the dataframe, but we can put conditions over here
df.loc[df['HP'] >= 60] #gives rows that has HP >=60
```

# Ways to Describe/Sort Data

```python
df.describe() #Descriptive statistics include those that summarize the central tendency, dispersion and shape of a dataset's distribution, excluding NaN values


df.sort_values(['Name', 'HP'], ascending=[False, True], axis=0) #Sorts Names as decending and HP as ascending
```

# Manipulating Data

```python
# adding a column
df['total']=df.HP+df.Attack
df

# dropping a column, df.drop() returns a new df with manipulated data
df=df.drop(columns='total')
df


# Using df.iloc() to add sum of all columns
df['total'] = df.iloc[:,4:10].sum(axis=1)  #axis 0 means along the rows(vertical) and axis 1 means along the column(horizontal)

df
```

# Writing the df to a csv

```python
#writing df to csv
df.to_csv('modified.csv', index=False, sep='\t') # here index is not specified will give index to each row, sep if not specified the demilliter will default be a ','
```

# Manipulating the DataFrame

```python
import re
#this list all rows that has Type 1 as fire, re.I means it will not consider the casing of string
df.loc[(df['Type 1'].str.contains('fire',flags=re.I,regex=True))]

# now loc() takes second argument that is the column we want to modify, in this case first it will locate all rows with type1 as fire, and to those rows the generation will be set to 7
df.loc[(df['Type 1'].str.contains('fire',flags=re.I,regex=True)),'Generation'] = 7
df
```
